### Size column have "Varies with device" value so removing them by using below command

```python
data = data[data['Size'].str.contains('Varies with device') == False]
data
```

### Multiply the value by 1,000, if size is mentioned in Mb

```python
data_M = data[data['Size'].str.contains('M')]
data_M
```

```python
data_M['Size'] = data_M['Size'].str.extract('(\d+)')
data_M
```

```python
data_M.dtypes
```

```python
data_M['Size'] = 1000*data_M['Size'].astype(int)
data_M
```

```python
data_M.dtypes
```

```python
data_K = data[data['Size'].str.contains('k')]
data_K['Size'] = data_K['Size'].str.extract('(\d+)')
data_K['Size'] = data_K['Size'].astype(int)
data_K
```

```python
new_data = pd.concat([data_M,data_K])
new_data = new_data.reset_index()
new_data
```

### Reviews is a numeric field that is loaded as a string field. Convert it to numeric (int/float)

```python
new_data['Reviews'] = new_data['Reviews'].astype(int)
```

### Installs field is currently stored as string and has values like 1,000,000+. Treat 1,000,000+ as 1,000,000 and remove ‘+’, ‘,’ from the field, convert it to integer

```python
new_data['Installs'] = new_data['Installs'].map(lambda x: str(x).replace('+','')) #x iterator replace the + with nothing
new_data['Installs'] = new_data['Installs'].map(lambda x: str(x).replace(',','')) #x iterator replace the , with nothing
new_data['Installs'] = new_data['Installs'].astype(int)
```

### Price field is a string and has $ Symbol. Remove ‘$’ sign, and convert it to numeric

```python
new_data['Price'] = new_data['Price'].map(lambda x: str(x).replace('$','')) #x iterator replace the $ with nothing
new_data['Price'] = new_data['Price'].astype(float)
```

### Average rating should be between 1 and 5 as only these values are allowed on the play store. Drop the rows that have a value outside this range

```python
new_data = new_data[(new_data['Rating'] > 1) & (new_data['Rating'] < 5)]
```

## Mean Squared Error

```python
X = new_data[['Rating']].values
Y = new_data[['Installs']].values
```

```python
# Split the Ratio into Training and Test Set
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.3,random_state = 12)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,Y_train)
```

```python
Y_pred = regressor.predict(X_test)
```

```python
# MSE
from sklearn.metrics import mean_squared_error
mean_squared_error(Y_test,Y_pred)
```

```python
# RMSE: Root Mean Squared Error
import math
math.sqrt(mean_squared_error(Y_test,Y_pred))
```

### Coefficient of Determination

```python
from sklearn.metrics import r2_score
r2_score(Y_test,Y_pred)
```
